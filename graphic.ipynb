```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('results.csv')

_, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))

ax1.plot(df['n'], df['area_n'], label=' ')
ax1.plot(df['n'], df['area_w'], label=' ')

ax1.axhline(y=0.25 * np.pi + 1.25 * np.arcsin(0.8) - 1.0, color='g', label=' ')

ax1.set_xlabel('  N')
ax1.set_ylabel('')
ax1.set_title('    ')
ax1.legend()
ax1.grid(True, alpha=0.3)

ax2.plot(df['n'], df['disp_n'], label=' ')
ax2.plot(df['n'], df['disp_w'], label=' ')

ax2.set_xlabel('  N')
ax2.set_ylabel(' ')
ax2.set_title('    ')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_yscale('log')

plt.tight_layout()
plt.savefig('monte_carlo_results.png', dpi=300, bbox_inches='tight')
plt.show()
```
